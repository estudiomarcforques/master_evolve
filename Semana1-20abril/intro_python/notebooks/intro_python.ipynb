## COMANDOS ÚTILES PARA LA CREACIÓN DE UN ENTORNO
* python -m venv venv     (se crea el entorno llamado env. El segundo env es el nombre que le queremos dar)

* venv\Scripts\activate       (activar el entorno)

* pip install pandas          (añadir bibliotecas)

* pip freeze > requirements.txt       (crear archivo requirements.txt con las librerias que necesita el entorno)

* pip install -r requirements.txt         (instalar todas las librerias que requiere el entorno con el archivo requirements creado)
