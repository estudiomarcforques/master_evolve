# 🖥️ Introducción a la Terminal (para Data Science)

Esta clase no va de memorizar comandos.
Va de **entender qué está pasando** cuando usas la terminal.

Piensa en la terminal como una conversación directa con tu ordenador:
- No hay botones
- No hay menús
- Solo instrucciones claras

> Si entiendes esto, el resto es vocabulario.

---
## 0. Preparación rápida (2–3 min)

Abre una terminal.

Si estás en **Windows**, usa **Git Bash** si lo has podido instalar.
Si no has podido, usa **PowerShell** con la chuleta de equivalencias.

Si estás en **Mac/Linux**, abre **Terminal**.

👉 Hoy yo voy a enseñar los comandos desde **Mac**, pero en **Git Bash** son iguales.
👉 Regla de oro: escribe comandos cortos, mira el resultado, y corrige.


---
## 1. Conceptos básicos (muy importante)

### 📁 Carpetas (directorios)
Son contenedores. Pueden tener:
- otras carpetas
- archivos

### 📄 Archivos
Son cosas concretas:
- un `.csv`
- un `.py`
- un `.txt`

La terminal **siempre está situada en una carpeta concreta**.

### ✅ Ejercicio 1 (3 min): entiende el lugar donde estás

1) Mira la barra o prompt de tu terminal: suele aparecer algo como tu usuario y una ruta.

2) Ahora ejecuta:
```
pwd
```

**Resultado esperado:** una ruta.

💡 Si sale algo raro o un error: copia exactamente el mensaje y revisa que has escrito bien.

---
## 2. ¿Dónde estoy ahora?

### `pwd`
**Print Working Directory**

Te dice la ruta exacta de la carpeta en la que estás.

Ejemplo:
```
/Users/alumno/Documentos
```

Traducción humana:
> "Estoy dentro de Documentos, que está dentro de usuario"

### ✅ Ejercicio 2 (3 min): localiza tu carpeta personal

Ejecuta:
```
cd ~
pwd
```

Pregunta rápida:
- ¿Tu ruta cambió?

👉 Esto es tu **home** (tu base de operaciones).

---
## 3. ¿Qué hay aquí?

### `ls`
Lista el contenido de la carpeta actual.

### Variantes útiles
- `ls -l` → lista detallada
- `ls -a` → incluye archivos ocultos

> Regla: nunca te muevas a ciegas. **Lista primero**.

### ✅ Ejercicio 3 (5 min): ver lo que hay y detectar ocultos

1) Ejecuta:
```
ls
```

2) Ejecuta:
```
ls -a
```

3) Busca algo que empiece por `.` (por ejemplo `.git`, `.ssh`, etc.)

💡 Idea: lo oculto no es “malo”; suele ser configuración.

---
## 4. Moverse entre carpetas

### `cd nombre_carpeta`
Entra en una carpeta.

### Atajos fundamentales
- `cd ..` → subir un nivel
- `cd ~` → ir a tu carpeta personal
- `cd -` → volver a donde estabas antes

### Truco profesional
Pulsa `TAB` para autocompletar nombres.

> Si escribes rutas largas a mano, estás trabajando mal.

### ✅ Ejercicio 4 (8 min): navegar sin perderte

1) Desde tu home (`cd ~`), elige una carpeta que veas en `ls`.

2) Entra con `cd` + TAB.

3) Comprueba:
```
pwd
ls
```

4) Vuelve atrás:
```
cd ..
```

5) Vuelve a la carpeta anterior usando:
```
cd -
```

💡 Si `cd -` no hace nada útil, es que no has cambiado de carpeta antes.

## 5. Crear cosas

### Crear carpetas
```
mkdir datos
```

### Crear archivos
```
touch notas.txt
```

Esto **no mete contenido**, solo crea el archivo.

### ✅ Ejercicio 5 (10 min): crea tu mini-proyecto

En tu home:

1) Crea una carpeta de trabajo:
```
mkdir terminal_practica
```

2) Entra:
```
cd terminal_practica
```

3) Crea estas carpetas:
```
mkdir data scripts output
```

4) Crea un archivo vacío:
```
touch README.txt
```

5) Verifica con:
```
ls
```

## 6. Copiar y mover (y renombrar)

### Copiar
```
cp origen destino
```

### Mover / renombrar
```
mv origen destino
```

💡 `mv` es el comando de “cambiar de sitio” y también el de “cambiar de nombre”.

### ✅ Ejercicio 6 (10 min): copia y renombra sin miedo

Dentro de `terminal_practica`:

1) Crea un archivo:
```
touch notas.txt
```

2) Cópialo:
```
cp notas.txt backup_notas.txt
```

3) Mueve `backup_notas.txt` a la carpeta `output`:
```
mv backup_notas.txt output/
```

4) Lista `output`:
```
ls output
```

## 7. Leer archivos de texto

### `cat archivo.txt`
Muestra el contenido completo.

### `less archivo.txt`
Permite leer poco a poco.

- Flechas para moverte
- `q` para salir

> En data science, leer archivos es parte del trabajo.

### ✅ Ejercicio 7 (8 min): escribe algo y léelo

Vamos a meter texto en un archivo (sin editor gráfico).

1) Ejecuta:
```
echo "Hola terminal" > saludo.txt
```

2) Léelo:
```
cat saludo.txt
```

3) Añade una línea más:
```
echo "Estoy aprendiendo a moverme" >> saludo.txt
```

4) Léelo con `less`:
```
less saludo.txt
```
Recuerda: `q` para salir.

## 8. Borrar cosas (con cuidado)

### Borrar un archivo
```
rm archivo.txt
```

### Borrar una carpeta
```
rm -rf carpeta
```

⚠️ Esto no va a la papelera.

> Si dudas, **no borres**.

### ✅ Ejercicio 8 (5 min): limpieza controlada

Dentro de `terminal_practica`:

1) Crea un archivo temporal:
```
touch borrar_esto.tmp
```

2) Comprueba que existe:
```
ls
```

3) Bórralo:
```
rm borrar_esto.tmp
```

4) Verifica que ya no está:
```
ls
```

## 9. Buscar archivos

### `find`
Busca cosas sin navegar a mano.

Ejemplos:
```
find . -name "*.csv"
find . -name "llave_sotano.txt"
```

Esto escala cuando hay muchas carpetas.


### ✅ Ejercicio 9 (8 min): buscar sin explorar

Desde `terminal_practica`:

1) Busca todos los `.txt`:
```
find . -name "*.txt"
```

2) Busca el archivo `saludo.txt`:
```
find . -name "saludo.txt"
```

Pregunta:
- ¿Qué significa el `.` en `find .`?
  - Pista: “desde aquí”.

## 10. Ejecutar Python desde la terminal

Si tienes un archivo `script.py`:
```
python3 script.py
```

En algunos equipos (sobre todo Windows con Git Bash) puede ser:
```
python script.py
```

Si falla:
- Python no está instalado
- Python no está en el PATH
- Estás usando un nombre de comando distinto (`python3` vs `python`)

La terminal **no adivina**, solo ejecuta lo que conoce.


### ✅ Ejercicio 10 (10 min): tu primer script ejecutado

Dentro de `terminal_practica/scripts`:

1) Entra:
```
cd scripts
```

2) Crea un archivo `hola.py` (opción simple):
```
echo "print('Hola desde Python')" > hola.py
```

3) Ejecútalo:
```
python3 hola.py
```

Si tu sistema usa `python` en vez de `python3`, prueba:
```
python hola.py
```

## 11. Misión práctica: el Laberinto (juego)

Vas a ejecutar un script que **crea un laberinto de carpetas** con pistas.

Objetivo: encontrar la llave, descubrir el código del candado y abrir el tesoro final.

### Pasos
1) Ve a la carpeta donde tienes `setup_juego.py`
2) Ejecuta:
```
python3 setup_juego.py
```
Si en tu sistema funciona `python` en vez de `python3`, usa:
```
python setup_juego.py
```
3) Entra en `Laberinto_Terminal`
4) Lee `leeme.txt`
5) Sigue las pistas hasta encontrar `llave_sotano.txt`
6) Descubre `CODIGO_COFRE`
7) Cuando llegues a `Caja_Fuerte`, ejecuta el comando exacto que te pida el candado
8) Ejecuta el tesoro final

Reglas:
- No usar explorador gráfico
- `pwd` + `ls` antes de moverte
- Usa TAB


### ✅ Ejercicio 11 (25–35 min): completar el laberinto

Checklist (marca mentalmente):
- [ ] He entrado en `Laberinto_Terminal`
- [ ] He leído `leeme.txt`
- [ ] He encontrado `PISTA_1.txt`, `PISTA_2.txt` y `PISTA_3.txt`
- [ ] He encontrado `llave_sotano.txt`
- [ ] He descubierto el código `DRAGON-404`
- [ ] He ejecutado `abrir_candado.py` con el código correcto
- [ ] He ejecutado `.tesoro/tesoro.py`

Pistas si te atascas:
- Usa `find . -name "PISTA_*.txt"`
- Usa `find . -name "llave_sotano.txt"`
- Usa `grep -R "CODIGO_COFRE" .`
- Si estás en `Caja_Fuerte`, lee `candado.txt`
- Si estás perdido: `pwd` y luego `ls`
