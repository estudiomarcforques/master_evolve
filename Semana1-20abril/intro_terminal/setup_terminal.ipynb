# 🛠️ Configuración de la Terminal (Para que todos tengamos la misma)

Uno de los problemas más comunes al aprender programación es que los comandos de **Windows** son diferentes a los de **Mac** y **Linux**.

Para evitar confusiones en clase y que todos sigamos las mismas instrucciones, vamos a unificar nuestros entornos.

---

## 🍎 Usuarios de Mac o Linux

**¡Buenas noticias! No tenéis que instalar nada.**

Vuestra terminal por defecto (zsh o bash) ya es perfecta para la clase.
Podéis abrirla ya mismo y comprobar que funciona con:

```
pwd
ls
```

---


## 🪟 Usuarios de Windows

La ruta principal para esta clase es **Git Bash**.

¿Por qué? Porque yo voy a enseñar los comandos desde **Mac**, y en **Git Bash** esos comandos son prácticamente iguales. Así todos podremos seguir la misma explicación sin traducir cada paso.

### Paso 1: Instalación (5 min)
1. Entra en **[git-scm.com/download/win](https://git-scm.com/download/win)**.
2. Haz clic en **"Click here to download"** para descargar el instalador.
3. Ejecuta el archivo descargado.
4. ⚠️ **IMPORTANTE:** Dale a **"Next"** a todas las opciones. No cambies nada. La configuración por defecto es suficiente.
5. Espera a que termine y cierra el instalador.


### Paso 2: Conectar VS Code con Git Bash

Ahora haremos que VS Code use esta nueva terminal por defecto.

1. Abre **Visual Studio Code**.
2. Presiona `Ctrl` + `Shift` + `P` para abrir la paleta de comandos.
3. Escribe y selecciona: **`Terminal: Select Default Profile`**.
4. En la lista que aparece, selecciona **`Git Bash`**.
5. Cierra la terminal actual y abre una nueva con `Ctrl` + `ñ`.

🎉 **Listo.** Si ves el símbolo `$` al final de la línea, ya estás en el entorno correcto para seguir los mismos comandos que yo mostraré en Mac.


---

## 🚑 Plan B: "No puedo instalar nada en este ordenador"

Si estás en un ordenador del trabajo o de la escuela y no puedes instalar Git Bash, usa **PowerShell** solo como plan B.

Importante:
- La clase está pensada para **Mac / Linux / Git Bash**.
- En **PowerShell** algunos comandos cambian.
- Si necesitas esa ruta, apóyate en el archivo **`Comandos terminal powershell.txt`**.

Equivalencias mínimas para no bloquearte:

| Acción | Comando del profesor (Mac / Git Bash) | Plan B en PowerShell |
| :--- | :--- | :--- |
| Ver dónde estoy | `pwd` | `pwd` |
| Listar archivos | `ls` | `ls` |
| Entrar / salir de carpetas | `cd carpeta` / `cd ..` | igual |
| Ver ocultos | `ls -a` | `Get-ChildItem -Force` |
| Crear archivo vacío | `touch archivo.txt` | `New-Item archivo.txt -ItemType File` |
| Ejecutar Python | `python3 archivo.py` | `python archivo.py` o `py archivo.py` |

Si puedes elegir, usa **Git Bash**.
