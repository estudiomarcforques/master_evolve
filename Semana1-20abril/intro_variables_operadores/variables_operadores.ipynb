## Tipos de datos nativos en python:

* int (enteros): números sin parte decimal (10, -5)

Ejemplo, si haces x = 10 estás creando una referencia llamada x que apunta a un objeto de tipo int con valor 10.

* float (flotantes): números con parte decimal (3.14, -1.0)

* str (cadenas de texto): secuencias de caracteres ("Hola", "123")

* bool (booleanos): valores lógicos True o False



```python
# Verificar tipos con type()
print(name, age, height, is_student)
print(type(name))      # str
print(type(age))        # int
print(type(height))      # float
print(type(is_student))  # bool
```

### Operadores Aritméticos

- **Suma**: `+`
- **Resta**: `-`
- **Multiplicación**: `*`
- **División**: `/` (devuelve float)
- **División entera**: `//` (descarta parte decimal)
- **Módulo**: `%` (resto de la división)
- **Exponente**: `**`


```python
a = 11
b = 3

print("Suma:", a + b)
print("Resta:", a - b)
print("Multiplicación:", a * b, type(a * b))
print("División:", a / b, type(a / b))
print("División redondeada:", round(a / b), type(round((a / b))))
print("División entera:", a // b)
print("Módulo:", a % b)
print("Exponente", a ** b)
```

```python
import math

print(math.sqrt(16))
```

### Operadores de Comparación

- **Igual que**: `==`
- **Distinto de**: `!=`
- **Mayor que**: `>` / **Menor que**: `<`
- **Mayor o igual**: `>=` / **Menor o igual**: `<=`

```python
x = 3
y = 4
z = 2
print("¿x es igual a y?", x == y)
print("¿x es diferente a y?", x != y)
print("¿x es mayor que y?", x > y)
print("¿x es menor o igual a y?", x <= y)
print("¿y está entre x y z", x < y < z)
```

### Operadores Lógicos

- `and`: verdadero si ambos operandos son verdaderos
- `or`: verdadero si al menos un operando es verdadero
- `not`: niega el valor de verdad


```python
a = True
b = False

print("a and b:", a and b)
print("a or b:", a or b)
print("not a:", not a)
```

Python es binario en espíritu (lo dejo como extra)

```python
# Comparaciones que muestran el comportamiento binario de los booleanos
print(f'False == 0?: {False == 0}')
print(f'True == 1?: {True == 1}')
print(f'True == 2?: {True == 2}')
print(f'False + True: {False + True}')
print(f'False - True: {False - True}')
print(f'True * 10: {True * 10}')
print(f'False * 10: {False * 10}')
```

## 🎟️ Ejercicio Integrador 2: Entrada al cine y descuento (sin condicionales)

El precio de la entrada del cine es **14€**.

✅ Las personas tienen un **descuento del 50%** si cumplen alguna de estas condiciones:

- **Menores de 12 años**.
- **Mayores de 65 años**.
- O si tienen una **tarjeta de cliente frecuente**.

---

### Datos:
- `age`: edad de la persona.
- `has_card`: `True` si tiene tarjeta de cliente frecuente, `False` si no.

---

### Actividades:

1. Calcula si **le corresponde el descuento** (muestra `True` o `False`).
2. Si quieres, calcula el precio final del boleto multiplicando el precio original por `0.5` si tiene descuento, o por `1` si no tiene descuento.  
   *(Sugerencia: puedes usar operadores aritméticos y lógicos, sin usar `if`).*

---

✅ Ejemplo de salida esperada:

> ¿Tiene descuento? True Precio final de la entrada: 7.50

```python
age= int(input("Introduce tu edad: "))

print(age)

input_tarjeta = input("¿Eres cliente? (si/no): ")

tarjeta_cliente = (input_tarjeta == "si")

print("¿Tienes tarjeta de cliente?", tarjeta_cliente)

descuento = age <= 12 or age >= 65 or tarjeta_cliente == True

precio = 14

print(f"¿Tienes descuento? {descuento}. El precio de las entradas es: {precio *(1-(0.5*descuento))}")
```

### 🆕 Métodos de String (Data Cleaning)
En Data Science, los datos nunca vienen limpios. Aquí unos trucos básicos para normalizar texto.

```python
correo = "   USUARIO@GMAIL.COM   "

print(f"Original: '{correo}'")
print(f"Sin espacios extra: '{correo.strip()}'")
print(f"Todo minúsculas: '{correo.strip().lower()}'")
print("\n")

name = "paula rodríguez simón"
print(f"Original: {name}")
print(f"Mayúsculas {name.upper()}")
print(f"Capitalizado primera palabra: {name.capitalize()}")
print(f"Capitalizado como título: {name.title()}")
```

### Tips importantes

- Python distingue entre mayúsculas y minúsculas en los nombres de variables.
- Las variables deben nombrarse con cuidado y evitar palabras reservadas (ej. `def`, `class`, `if`, `for`, etc.).
- Se recomienda usar nombres descriptivos para mejorar la legibilidad del código.

```python
# Nombres descriptivos y buenas prácticas
temperatura_celsius = 22.5
esta_nublado = False

print("La temperatura es:", temperatura_celsius, "°C")
print("¿Está nublado?", esta_nublado)
```

### Reto 1: Calculadora de IMC
Pide por `input` el peso y la altura de un alumno y muestra su IMC con 2 decimales.

```python
height= float(input("Indica tu altura en cm: "))

weight= float(input("Indica tu peso en kg: "))

imc = weight / (height/100)**2

print(f"Tu IMC es: {imc:.2f}")
print(f"Tu IMC es: {round(imc,2)}")

```
